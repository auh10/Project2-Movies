```python
movie_title = []
title_name_elem = soup.find_all('h3',{'class':'lister-item-header'})
for item in title_name_elem:
    movie_title.append(item.find('a').text)
len(movie_title)
```

```python
film_rating = []

film_rating_name_elem = soup.find_all(class_='certificate')
for item in film_rating_name_elem:
    film_rating.append(item.text)
    
len(film_rating)
film_rating_df = film_rating
```

```python
year = []
year_name_elem = soup.find_all('span',{'class':'lister-item-year text-muted unbold'})
for item in year_name_elem:
    year.append(item.get_text().strip("()"))

year
```

```python
duration = []

duration_name_elem = soup.find_all(class_='runtime')
for item in duration_name_elem:
    duration.append(item.text.strip("min"))
    
duration
```

```python
genre = []

genre_name_elem = soup.find_all(class_='genre')
for item in genre_name_elem:
    genre.append(item.text.strip())
    
genre
```

```python
meta_score = []
score_name_elem = soup.find_all('div',{'class':'inline-block ratings-metascore'})
for item in score_name_elem:
    meta_score.append(item.find('span').text.strip())
len(meta_score)

meta_score_df = meta_score
```
